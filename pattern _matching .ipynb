```python
import spacy
nlp = spacy.load('en_core_web_sm')
def ADJ_NOUN(input):
    processed_input = nlp(input)

    matched_parts = []

    for tok in processed_input:
        matched_part = ''
        # if the word's POS tag is Noun and
        # dependency tag is a kind of subject or object
        if (tok.pos_ == 'NOUN') \
                and \
                (tok.dep_ in
                 ['dobj', 'pobj', 'nsubj', 'nsubjpass'])\
                :

            # search among all nearby token
            for nearby_tok in tok.children:
                # if the word's POS tag is adjective
                # or dependency tag is compound
                if (nearby_tok.pos_ == 'ADJ') \
                        or \
                        (nearby_tok.dep_ == 'compound')\
                        :
                    matched_part += nearby_tok.text + ' '

            if len(matched_part) != 0:
                matched_part += tok.text

        if len(matched_part) != 0:
            matched_parts.append(matched_part)

    return matched_parts

h=ADJ_NOUN("Agricultural production depends on technology.")
print(h)
```

```python
def N_V_N(input):
    processed_input = nlp(input)

    matched_parts = []

    for tok in processed_input:

        # if a verb is found
        if (tok.pos_ == 'VERB'):

            matched_part = ''

            # extract NOUN or PROPN
            for nearby_tok in tok.lefts:

                if (nearby_tok.dep_ in ['nsubj', 'nsubjpass']) \
                    and \
                    (nearby_tok.pos_ in ['NOUN', 'PROPN', 'PRON']) \
                    :

                    # subject added
                    matched_part += nearby_tok.text

                    # lemma of the verb added
                    matched_part += ' ' + tok.lemma_

                    # find object
                    for nearby_tok in tok.rights:

                        # object added
                        if (nearby_tok.dep_ in ['dobj']) \
                            and \
                            (nearby_tok.pos_ in ['NOUN', 'PROPN']) \
                                :
                            matched_part += ' ' + nearby_tok.text
                            matched_parts.append(matched_part)

    return matched_parts


h = N_V_N("The sun provides energy to animals")
print(h)
```
